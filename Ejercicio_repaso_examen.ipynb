```python
import sys
import numpy as np

entradas = []
ubicaciones_disponibles = list(range(1, 51))
ganancias_totales = 0

def menu():
    print("1. Comprar entradas")
    print("2. Mostrar ubicaciones disponibles")
    print("3. Ver listado de asistentes")
    print("4. Mostrar ganancias totales")
    print("5. Salir")

def comprar_entradas():
    cantidad = int(input("\nIngrese la cantidad de entradas que desea (máximo 2 entradas): "))
    if cantidad > 2:
        print("\nSolo dos entradas como máximo.\n")
        return

    print("\nUbicaciones disponibles:\n")
    mostrar_ubicaciones_disponibles()

    for i in range(cantidad):
        asiento = int(input("\nElija el asiento que desea: "))
        if asiento not in ubicaciones_disponibles:
            print("\nAsiento no disponible.\n")
            return

        if asiento <= 20:
            precio = 100000
        elif asiento <= 30:
            precio = 50000
        elif asiento <= 50:
            precio = 10000

        run = input("\nIngrese su rut (sin dígito verificador): ")
        entrada = np.array([run, asiento, precio])
        entradas.append(entrada)

        ubicaciones_disponibles.remove(asiento)
        global ganancias_totales
        ganancias_totales += precio

    print("\nCompra realizada con éxito.\n")

def mostrar_ubicaciones_disponibles():
    print("----------------------------------ESCENARIO----------------------------------")
    for i in range(1, 51):
        if i in ubicaciones_disponibles:
            print(i, end="\t")
        else:
            print("X", end="\t")

        if i % 10 == 0:
          print()


def listado_asistentes():
    if len(entradas) == 0:
        print("\nNo hay entradas registradas.\n")
    else:
        for entrada in entradas:
            print("\nRun:", entrada[0])
            print("Asiento N°:", entrada[1])
            print("Precio de la entrada:", entrada[2])
            print()

def mostrar_ganancias_totales():
    print(f"\nLas ganancias totales de la venta de entradas es de: ${ganancias_totales}")

def salir():
    print("\nGracias por comprar con nosotros.")
    print("Javier Silva - V1 - 02/07/2023\n")
    sys.exit()

while True:
    menu()
    opcion = input("\nElija una opción: ")
    if opcion == "1":
        comprar_entradas()
    elif opcion == "2":
        mostrar_ubicaciones_disponibles()
    elif opcion == "3":
        listado_asistentes()
    elif opcion == "4":
        mostrar_ganancias_totales()
    elif opcion == "5":
        salir()
    else:
        print("\nIngrese una opción válida.\n")

```

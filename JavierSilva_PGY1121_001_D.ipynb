```python
import numpy as np
import sys

entradas = []
ubicaciones_disponibles = list(range(1,101))
ganancias_totales = []
total = 0
suma_entradas = 0

def menu():
  print("1. Comprar entradas")
  print("2. Mostrar ubicaciones disponibles")
  print("3. Ver listado de asistentes")
  print("4. Mostrar ganancias totales")
  print("5. Salir")

def comprar_entradas():
  cantidad = int(input("\nCuántas entradas desea (máximo 3): "))
  if cantidad > 3:
    print("\nLa cantidad de entradas no puede ser mayor a 3.\n")
    return

  print("\nUbicaciones disponibles: ")
  mostrar_ubicaciones_disponibles()

  for i in range(cantidad):
    asiento = int(input("\nElija un asiento: "))
    if asiento not in ubicaciones_disponibles:
      print("Asiento no disponible.")
      return

    if asiento <= 20:
      precio = 120000
      tipo_entrada = "Platinum"
    elif asiento <= 50:
      precio = 80000
      tipo_entrada = "Gold"
    elif asiento <= 100:
      tipo_entrada = "Silver"
      precio = 50000

  run = int(input("\nIngrese su rut para completar el registro: "))
  entrada = np.array([run, asiento, precio])
  entradas.append(entrada)
  ubicaciones_disponibles.remove(asiento)

  global total
  total += precio

  ganancias = np.array([tipo_entrada, cantidad, precio])
  ganancias_totales.append(ganancias)

  global suma_entradas
  suma_entradas += cantidad
  print("\nCompra de entradas realizada con éxito.\n")

def mostrar_ubicaciones_disponibles():
  print("----------------------------------ESCENARIO----------------------------------")
  for i in range(1,101):
    if i in ubicaciones_disponibles:
      print(i, end="\t")
    else:
      print("X", end="\t")

    if i % 10 == 0:
      print()

def listado_asistentes():
  if len(entradas) == 0:
    print("\nNo hay asistentes registrados.\n")

  print("\nLISTA DE ASISTENTES POR RUT Y ASIENTO\n")
  for entrada in entradas:
    sorted(entrada)
    print(entrada)

  print()

def mostrar_ganancias_totales():
  if len(ganancias_totales) == 0:
    print("\nNo hay ganancias registradas.\n")

  print("\nTIPO DE ENTRADA Y CANTIDAD DE ENTRADAS\n")
  for ganancias in ganancias_totales:
    print(ganancias)

  print(f"\nCantidad total de entradas: {suma_entradas}\nTotal de todas las entradas: ${total}\n")


def salir():
  print("Gracias por comprar con nosotros.")
  print("Javier Silva - V1 - 12/07/2023")
  sys.exit()

while True:
  menu()
  opcion = input("\nElija una opción: ")
  if opcion == "1":
    comprar_entradas()
  elif opcion == "2":
    mostrar_ubicaciones_disponibles()
  elif opcion == "3":
    listado_asistentes()
  elif opcion == "4":
    mostrar_ganancias_totales()
  elif opcion == "5":
    salir()
  else:
    print("\nSeleccione una opción válida.\n")
```

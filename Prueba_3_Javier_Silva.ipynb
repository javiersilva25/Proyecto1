```python
import sys

lista_clientes = []

def menu():
  print("------MENÚ------")
  print("1. Grabar datos de un cliente")
  print("2. Buscar cliente por rut")
  print("3. Imprimir certificados")
  print("4. Salir del programa")

def datos_cliente():
  run_cliente = input("Ingrese su RUN (con guión): ")

  nombre_cliente = input("Ingrese su nombre: ")

  while len(nombre_cliente) < 8:
    print("El nombre debe tener al menos 8 caracteres.")
    nombre_cliente = input("Ingrese su nombre: ")


  edad_cliente = int(input("Ingrese su edad: "))

  while edad_cliente < 0:
    print("La edad no puede ser menor a 0.")
    edad_cliente = int(input("Ingrese su edad: "))


  if edad_cliente > 0:
    clientes = {
        'run_cliente': run_cliente,
        'nombre_cliente': nombre_cliente,
        'edad_cliente': edad_cliente
    }

    lista_clientes.append(clientes)
    print("\nCliente ingresado con éxito.\n")

def buscar_cliente():
  run_cliente = input("Ingrese el RUN (con guión): ")
  encontrado = False

  for cliente in lista_clientes:
    encontrado = True
    if cliente['run_cliente'] == run_cliente:
      print("\nRUN:", cliente['run_cliente'])
      print("Nombre:", cliente['nombre_cliente'])
      print("Edad:", cliente['edad_cliente'])
      print(" ")

  if encontrado == False:
    print("Cliente no encontrado.")

def certificados_cliente():
  encontrado = False
  for cliente in lista_clientes:
    encontrado = True
    print("Certificado de nacimiento:\n")
    print("RUN:", cliente['run_cliente'])
    print("Nombre:", cliente['nombre_cliente'])
    print("Edad:\n", cliente['edad_cliente'])
    print(" ")

    print("Certificado de estado conyugal:\n")
    print("RUN:", cliente['run_cliente'])
    print("Nombre:", cliente['nombre_cliente'])
    print("Edad:", cliente['edad_cliente'])
    print("Estado civil: Viudo/a")
    print(" ")

  if encontrado == False:
    print("\nNo hay clientes registrados.\n")

def salir():
  print("\nGracias por usar el programa.")
  print("Javier Silva - V1\n")
  sys.exit()

while True:

  menu()
  opcion = input("Elija una opción: ")

  if opcion == "1":
    datos_cliente()
  elif opcion == "2":
    buscar_cliente()
  elif opcion == "3":
    certificados_cliente()
  elif opcion == "4":
    salir()
  else:
    print("\nIngrese una opción válida.\n")









```
